```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
depth_data = np.load('/home/scilab/teleoperation/avp_teleoperate/teleop/utils/data/episode_0010/depths/000105_depth_0.npy')
print(depth_data.shape)
print(depth_data.dtype)
print(depth_data)
print(depth_data.min(), depth_data.max())
zero_indices = np.argwhere(depth_data == 0)

print(f"Zero‐valued pixels: {zero_indices.shape[0]}")
zero_mask = (depth_data == 0)

```

```python
plt.figure(figsize=(6,4))
plt.imshow(zero_mask, cmap='gray')
plt.title("Zero‐Value Pixel Mask")
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.imshow(depth_data, cmap='jet')   
plt.title("Depth")
plt.axis("off")
plt.colorbar(label="uint8 value")
plt.show()
```
