```python
import cv2
import numpy as np
def draw_distance_gauge(image, distance, max_distance=0.3):  # max_distance: 30cm
    h, w = image.shape[:2]
    bar_width = 300
    bar_height = 30

    # 거리 비율 (0~1)
    ratio = max(0, min(1, 1 - distance / max_distance))

    # 게이지 바 배경
    bar_x, bar_y = 50, h - 50
    cv2.rectangle(image, (bar_x, bar_y), (bar_x + bar_width, bar_y + bar_height), (50, 50, 50), -1)

    # 채워진 게이지
    filled_width = int(bar_width * ratio)
    cv2.rectangle(image, (bar_x, bar_y), (bar_x + filled_width, bar_y + bar_height), (0, 255, 0), -1)

    # 텍스트 표시
    cv2.putText(image, f"Distance: {distance*100:.1f} cm", (bar_x, bar_y - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

    return image

img = cv2.imread('/home/scilab/Documents/teleoperation/images/1.jpg')

# 거리 게이지 표시
image = draw_distance_gauge(img, distance=0.1, max_distance=0.3)  # 10cm로 설정

# 이미지 보여주기
cv2.imshow("Distance Gauge", image)
cv2.waitKey(0)   # 키 입력 대기
cv2.destroyAllWindows()
```
